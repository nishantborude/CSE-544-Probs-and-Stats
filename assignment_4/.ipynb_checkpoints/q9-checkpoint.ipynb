### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

### Reading data into data1 (sigma3) and data2 (sigma100)

```python
data1 = pd.read_csv('q9_sigma3.dat', header=None)
data2 = data_2 = pd.read_csv('q9_sigma100.dat', header=None)
```

### Define the function that takes sigma and data and gives mean and variance of the posterior for each iteration

```python
def funct(sigma, data):
    a = 0
    b_2 = 1
    a_list = []
    b_list = []
    se_2 = (sigma**2) / data.shape[1]
    
    for i in range(data.shape[0]):
        pos_data = data.iloc[i, :]
        X_bar = np.mean(pos_data)
        x = (b_2 * X_bar + se_2 * a) / (b_2 + se_2)
        y_2 = (b_2 * se_2) / (b_2 + se_2)
        a = x
        b_2 = y_2
        a_list.append(a)
        b_list.append(b_2)

    plt.figure(figsize=(30,20))
    for i in range(len(a_list)):
        mu = a_list[i]
        variance = b_list[i]
        sigma = np.sqrt(variance)
        x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
        plt.plot(x,mlab.normpdf(x, mu, sigma))
    
    leg_list= []
    for i in range(len(a_list)):
        leg_list.append('Row{}-Mean {:.4g}, Variance {:.3g}'.format(i + 1, a_list[i], b_list[i]))
    plt.legend(leg_list, prop={'size': 25})
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    plt.title("Posterior Plot", fontsize = 25)
    plt.xlabel("Distribution", fontsize = 25)
    plt.ylabel("Probability", fontsize = 25)
    plt.savefig('a.jpg', bbox_inches='tight', pad_inches = 1)
    plt.show()
    
    d = pd.DataFrame()
    d[0] = a_list
    d[1] = b_list
    d.columns = ['Mean', 'Variance']
    return d
```

### Plot the graph of posterior for every iteration for sigma = 3

```python
d3 = funct(3, data1)
print(d3)
```

### Plot the graph of posterior for every iteration for sigma = 100

```python
d100 = funct(100, data2)
print(d100)
```
