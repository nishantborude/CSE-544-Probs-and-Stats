```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def one_d_data(x):
    y = [0]
    y_tmp = 0
    step = 1/len(x)
    for i in x:
        y_tmp += step
        y.append(y_tmp)
    return y
```

```python
def plot_fn(x, y):
    plt.figure(figsize = (20,20))
    plt.xticks(x)
    plt.grid(True)
    plt.step(x, y, where='post')
    
    plt.show()
```

```python
def two_d_data(x):
    fn_mat = np.zeros((len(x), 99))
    
    m = len(x) 
    step = 1/len(x[0])
    
    for i in range(len(x)):
        tmp = 0
        prev = 0
        for j in range(len(x[0])):
            for k in range(prev, x[i][j]):
                fn_mat[i][k] = tmp
            tmp += step
            prev = x[i][j]
        for k in range(prev, 99):
            fn_mat[i][k] = tmp
            
    fn_avg = np.array([np.array(y) for y in fn_mat])
    return fn_avg.mean(0)
```

```python
x_10 = np.random.randint(1, 100 ,size = 10)
x_10.sort()

y = one_d_data(x_10)

x_10 = np.insert(x_10, 0, 0)

plot_fn(x_10, y)
```
